```python
#Added up the total cost of montly ultilities and bills for a single person 
#Assumes the individual owns a car, doesnt own a home but rents,
#Assumes the individual lives in one bed room apartment, 250 mins of calls a month, goes to the gym
#Assumes the individual uses 125 literes of gasoline
#Assumes individual not living in the city centre
cost_of_living['single_person_month']= (
    (cost_of_living['one_bed_apt_outside_centre']) +
    (cost_of_living['utilities_month']) +
    (cost_of_living['gasoline_liter']*125) +
    (cost_of_living['monthly_internet']) +
    (cost_of_living['call_per_min']*250) + 
    (cost_of_living['monthly_gym']) 
)

#Calculated cost of utilaties for family of four
#Assumes the family rents a 3 bedroom apartment, pays 15% more for utilites, pays 20% more for gasoline,
#Assumes the family has 2 children with one going to international school and the other to nursery,
#Assumes both parents have a gym membership
#Assumes the family uses 125 literes of gasoline a month
#Assumes family not living in the city centre
cost_of_living['family_month']= (
    (cost_of_living['three_bed_apt_outside_centre']) +
    (cost_of_living['utilities_month']*1.15) +
    (cost_of_living['gasoline_liter']*125) +
    (cost_of_living['monthly_internet']) +
    (cost_of_living['call_per_min']*275) +
    (cost_of_living['monthly_gym']*2) +
    (cost_of_living['nursery_per_year']/12)   
)
cost_of_living.head(30)
```
